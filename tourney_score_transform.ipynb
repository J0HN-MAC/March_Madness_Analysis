```python
import pandas as pd
```

```python
# Read in CSV. Name of dataframe is meant to correct  incorrect name of 
tourney_score_df = pd.read_csv('Resources/Tournament Team Data.csv')
tourney_score_df
```

```python
# Drop rows with 2023 data (games not yet played)
clean_score_df = tourney_score_df[tourney_score_df["YEAR"] != 2023]
clean_score_df
```

```python
# Create a new "team" column that combines values from "YEAR" and "TEAM" columns
clean_score_df['team'] = clean_score_df['YEAR'].astype(str) + ' ' + clean_score_df['TEAM'].astype(str)
clean_score_df.head()
```

```python
# Only keep "team", "CURRENT ROUND", "SEED", "SCORE" columns
clean_score_df = clean_score_df[["team", "CURRENT ROUND", "SEED", "SCORE"]]
clean_score_df
```

```python
# Rename the columns
clean_score_df.rename(columns = {'CURRENT ROUND':'current_round', 'SEED':'seed',
                              'SCORE':'score'}, inplace = True)
clean_score_df.head()
```

```python
# Add "index" label to index column so this table will have a primary key
clean_score_df.index.name = "index"
clean_score_df.head()
```

```python
# Export datafrome to csv file
file_path = 'Resources/clean_score_data.csv'
clean_score_df.to_csv(file_path)
```
