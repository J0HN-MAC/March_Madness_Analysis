```python
import pandas as pd
```

```python
# Read in CSV
team_df = pd.read_csv('Resources/Tournament Game Data.csv')
team_df
```

```python
# Rename TEAM column to school
clean_team_df = team_df.rename(columns={'TEAM': 'school'})
clean_team_df.head()
```

```python
# Create a new "team" column that combines values from "YEAR" and "school" columns
clean_team_df['team'] = clean_team_df['YEAR'].astype(str) + ' ' + clean_team_df['school'].astype(str)
clean_team_df.head()
```

```python
#Drop round of 68 teams
clean_team_df = clean_team_df[clean_team_df['ROUND'] != 68]
clean_team_df
```

```python
clean_team_df['ROUND'].dtypes
```

```python
# Replace values in "ROUND" column to convert round into tournament wins
clean_team_df.ROUND.replace([64, 32, 16, 8, 4, 2, 1], [0, 1, 2, 3, 4, 5, 6], inplace=True)

# Rename "ROUND" column to "tournaments_wins"
clean_team_df = clean_team_df.rename(columns={'ROUND': 'tournament_wins'})

clean_team_df.head()
```

```python
# Only keep "team", "CURRENT ROUND", "SEED", "SCORE" columns
clean_team_df = clean_team_df[["team", "KENPOM ADJUSTED EFFICIENCY", "KENPOM ADJUSTED OFFENSE", "KENPOM ADJUSTED DEFENSE", "SEED", "school", "YEAR", "tournament_wins"]]
clean_team_df.head()
```

```python
# Rename the columns
clean_team_df.rename(columns = {'KENPOM ADJUSTED EFFICIENCY':'kp_adj_eff', 'KENPOM ADJUSTED OFFENSE':'kp_adj_off',
                              'KENPOM ADJUSTED DEFENSE':'kp_adj_def', 'SEED':'seed', 'YEAR':'year',}, inplace = True)
clean_team_df.head()
```

```python
# Export datafrome to csv file
file_path = 'Resources/clean_team_data.csv'
clean_team_df.to_csv(file_path, index=False)
```
